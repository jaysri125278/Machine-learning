# Gradient Descent

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 3, 5, 7, 11])

# Parameters
alpha = 0.001  # Learning rate
iterations = 1000 # Number of iterations
cost_values = []
m_values = []

# Initialize parameters
c = 0
m = 0

n = len(x)

# Gradient Descent
for _ in range(iterations):
    # Compute the predictions
    y_pred = c + m * x

    # Compute the cost function (optional, for monitoring)
    cost = (1 / (2 * n)) * np.sum((y_pred - y) ** 2)

    # Compute the gradients
    d_c = (1 / n) * np.sum(y_pred - y)
    d_m = (1 / n) * np.sum((y_pred - y) * x)

    # Update the parameters
    c -= alpha * d_c
    m -= alpha * d_m

    cost_values.append(cost)
    m_values.append(m)

    # (Optional) Print cost for every 100 iterations
    if _ % 100 == 0:
        print(f"Iteration {_}: Cost {cost}, c {c}, m {m}")

# Final parameters
print(f"Final parameters: c = {c}, m = {m}")

# Plot the data and the regression line
plt.scatter(x, y, color='blue', label='Data points')
plt.plot(x, c + m * x, color='red', label='Regression line')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
plt.plot(m_values, cost_values, label='Cost Function')
plt.xlabel('Slope (m)')
plt.ylabel('Cost Function')
plt.title('Slope (m) vs. Cost Function')
plt.legend()
plt.show()
```

```python
min_cost_index = np.argmin(cost_values)
min_cost = cost_values[min_cost_index]
optimal_m = m_values[min_cost_index]

print(f"Global Minima at Iteration {min_cost_index+1}:")
print(f"Optimal Slope (m) = {optimal_m}")
print(f"Optimal Intercept (c) = {c}")
print(f"Minimum Cost = {min_cost}")

# Plot the graph between m and cost function values
plt.plot(m_values, cost_values, label='Cost Function')
plt.axhline(y=min_cost, color='r', linestyle='--', label=f'Min Cost: {min_cost:.4f}')
plt.axvline(x=optimal_m, color='g', linestyle='--', label=f'Optimal m: {optimal_m:.4f}')
plt.xlabel('Slope (m)')
plt.ylabel('Cost Function')
plt.title('Slope (m) vs. Cost Function')
plt.legend()
plt.show()
```

# Ordinary Least Squares (OLS)



```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 3, 5, 7, 11])

# Calculate the means
x_mean = np.mean(x)
y_mean = np.mean(y)

# Calculate the slope (m)
numerator = np.sum((x - x_mean) * (y - y_mean))
denominator = np.sum((x - x_mean) ** 2)
min_cost_index = numerator / denominator

# Calculate the intercept (c)
c = y_mean - m * x_mean

print(f"Slope (beta_1): {m}")
print(f"Intercept (beta_0): {c}")

# Plot the data and the regression line
plt.scatter(x, y, color='blue', label='Data points')
plt.plot(x, c + m * x, color='red', label='Regression line')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
